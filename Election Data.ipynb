### FOR LOOP FOR FILTERING DATA BASED ON PARTY

```python
import numpy as np
three=[]
two=[]
one=[]



for i in range(0,len(dd_2d),3):                       #based on place value of each element
    if (dd_2d[i,2]==dd_2d[i+1,2]==dd_2d[i+2,2]):
        three.append(dd_2d[i])
        three.append(dd_2d[i+1])
        three.append(dd_2d[i+2])
    #elif (dd_2d[i,2]!=dd_2d[i+1,2]!=dd_2d[i,2]!=dd_2d[i+2,2]):
    elif ((dd_2d[i,2] == dd_2d[i+1,2]) or (dd_2d[i,2] == dd_2d[i+2,2]) or (dd_2d[i+1,2] == dd_2d[i+2,2])):
        two.append(dd_2d[i])
        two.append(dd_2d[i+1])
        two.append(dd_2d[i+2])
    else:
        one.append(dd_2d[i])
        one.append(dd_2d[i+1])
        one.append(dd_2d[i+2])
```
